**Find if there is any relationship between the categorical variables and the output
variable in the data.**

since you want to find relationships between categorical variables and the product purchased, it’s advisable to convert age(continuous) into categories for clarity in visualizations.

```python
bins = [0, 18, 25, 35, 45, 55, 65, 100]
labels = ['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '66+']

# Convert Age to categorical
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

```

```python
# Categorical variables to analyze
categorical_columns = ['MaritalStatus', 'Gender', 'AgeGroup']


plt.figure(figsize=(15, 10))

# Loop through categorical columns and create count plots
for i, column in enumerate(categorical_columns, 1):
    plt.subplot(2, 2, i)
    sns.countplot(x=column, hue='Product', data=df, palette='viridis')
    plt.title(f'Relationship between {column} and Product Purchased')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

```

**Observations:**

**Marital Status:**
Single and partnered individuals favor KP281 the most.
Partnered individuals have a significantly higher preference for KP281 compared to single individuals.

**Gender:**
Both males and females prefer KP281, but males show a much higher count across all products.
Females show a relatively lower preference for KP781 compared to males.

**Age Group:**
The 19-25 and 26-35 age groups are the most active buyers, especially for KP281.
Purchases drop significantly for older age groups, especially for KP781.

**INSIGHTS:**
KP281 is the top choice across all segments, with partnered individuals and males showing the highest preference.
KP781 has low purchase counts among females and older age groups (45+).
Younger buyers (19-35) drive the majority of purchases for all products, particularly KP281.

**Recommendations:**
Target KP281 with promotions focused on males, partnered individuals, and younger age groups (19-35) to maintain sales momentum.
Boost KP781 sales by designing campaigns aimed at females and older customers, with tailored offers or product features.
Use age-specific and demographic-specific marketing to engage less active segments like older age groups and females to diversify the product reach.

**Find the marginal probability (what percent of customers have purchased
KP281, KP481, or KP781)**

```python
product_probabilities = pd.crosstab(index=df['Product'], columns='Count', normalize=True) *100

product_probabilities.columns = ['Marginal Probability (%)']

print(product_probabilities)

```

**Find the probability that the customer buys a product based on each column.**

```python

columns = ['Age', 'Gender', 'Education', 'MaritalStatus', 'Usage', 'Fitness', 'Income', 'Miles']

probabilities = {}
for col in columns:
    crosstab_result = pd.crosstab(index=df[col], columns=df['Product'], normalize='index') * 100

    probabilities[col] = crosstab_result
for col, prob_df in probabilities.items():
    print(f"\nProbabilities based on {col}:\n")
    print(prob_df)

```
